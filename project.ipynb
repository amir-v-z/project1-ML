<h3 align="right"  style="color: #ff6b35">(Histogram , Boxplot نمودار های) نمایش گرافیکی از توزیع و پراکندگی ستون‌های عددی</h3>

```python
numeric_cols = ['Age', 'Income', 'AdSpend', 'ClickThroughRate', 'ConversionRate',
                'WebsiteVisits', 'PagesPerVisit', 'TimeOnSite', 'SocialShares',
                'EmailOpens', 'EmailClicks', 'PreviousPurchases', 'LoyaltyPoints']

fig, axes = plt.subplots(13, 2, figsize=(12, 35))
axes = axes.flatten()

plot_idx = 0
for col in numeric_cols:
    
    # Histogram
    sns.histplot(df[col], kde=True, color="#ff6b35", ax=axes[plot_idx])
    plot_idx += 1
    
    # Boxplot
    sns.boxplot(y=df[col], color="#4cd137", ax=axes[plot_idx])
    axes[plot_idx].set_xlabel(col)
    axes[plot_idx].set_ylabel('Count')
    plot_idx += 1

fig.text(0.24, 0.95, 'Histogram', ha='center', va='center', fontsize=18, fontweight='bold')
fig.text(0.80, 0.95, 'Boxplot', ha='center', va='center', fontsize=18, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.show()
```

<h3 align="right"  style="color: #4cd137">(Outlier Data Management) مدیریت نقاط پرت</h3>

<h3 align="right"  style="color: #ff6b35">EmailClicks شناسایی و مدیریت نقاط پرت در ستون</h3>

<h3 align="right"  style="color: #D980FA">قبل از مدیریت نقاط پرت‌</h3>

```python
clean_data2["EmailClicks"].count()
```

```python
q1 = clean_data2["EmailClicks"].quantile(0.25)
q3 = clean_data2["EmailClicks"].quantile(0.75)
IQR = q3 - q1

lower_whisker = q1 - 1.5 * IQR
upper_whisker = q3 + 1.5 * IQR

outliers = clean_data2[(clean_data2["EmailClicks"] < lower_whisker) |
                       (clean_data2["EmailClicks"] > upper_whisker)]

print(f"تعداد نقاط پرت: {outliers['EmailClicks'].count()}")
```

<h3 align="right"  style="color: #D980FA">بعد از مدیریت نقاط پرت‌</h3>

```python
clean_data3 = clean_data2.copy()
clean_data3.loc[outliers.index, "EmailClicks"] = np.nan

# Or with the mask method

# clean_data3["EmailClicks"] = clean_data3["EmailClicks"].mask(
#     (clean_data3["EmailClicks"]<lower_whister) | (clean_data3["EmailClicks"]>upper_whister),
#     np.nan
# )

clean_data3['EmailClicks'].count()
```

<h3 align="right"  style="color: #ff6b35">Income شناسایی و مدیریت نقاط پرت در ستون</h3>

<h3 align="right"  style="color: #D980FA">قبل از مدیریت نقاط پرت‌</h3>

```python
plt.figure(figsize=(5, 3))
sns.boxplot(x="Gender", y="Income", data=clean_data3, color="#0652DD")
plt.title("Income comparison by gender (with outliers)", fontsize=13, fontweight='bold')
plt.show()
```

```python
clean_data3["Income"].count()
```

```python
q1 = clean_data3["Income"].quantile(0.25)
q3 = clean_data3["Income"].quantile(0.75)
IQR = q3 - q1

lower_whisker = q1 - 1.5 * IQR
upper_whisker = q3 + 1.5 * IQR

outliers = clean_data3[(clean_data3["Income"] < lower_whisker) |
                       (clean_data3["Income"] > upper_whisker)]

print(f"تعداد نقاط پرت: {outliers['Income'].count()}")
```

<h3 align="right"  style="color: #D980FA">بعد از مدیریت نقاط پرت‌</h3>

```python
clean_data4 = clean_data3.copy()

clean_data4["Income"] = clean_data4["Income"].mask(
    (clean_data4["Income"]<lower_whisker) | (clean_data4["Income"]>upper_whisker),
    np.nan
)

clean_data4['Income'].count()
```

```python
plt.figure(figsize=(5, 3))
sns.boxplot(x="Gender", y="Income", data=clean_data4, color="#0652DD")
plt.title("Income comparison by gender (with outliers)", fontsize=13, fontweight='bold')
plt.show()
```

<h3 align="right"  style="color: #4cd137">(Feature Engineering) مهندسی ویژگی</h3>

<h3 align="right"  style="color: #ff6b35">به گروه‌های رده سنی معنادار Age تبدیل ستون</h3>

```python
age_group = [
    clean_data4['Age'] <= 18,
    (clean_data4['Age'] >= 19) & (clean_data4['Age'] <= 24),
    (clean_data4['Age'] >= 25) & (clean_data4['Age'] <= 34),
    (clean_data4['Age'] >= 35) & (clean_data4['Age'] <= 44),
    (clean_data4['Age'] >= 45) & (clean_data4['Age'] <= 54),
    (clean_data4['Age'] >= 55) & (clean_data4['Age'] <= 64),
    clean_data4['Age'] >= 65
]

labels = ['-18', '19-24', '25-34', '35-44', '45-54', '55-64', '65+']

clean_data4['AgeGroup'] = np.select(age_group, labels, default='Unknown')

clean_data4['AgeGroup'].value_counts(dropna=False).sort_index()
```

<h3 align="right"  style="color: #4cd137">پر کردن مقادیر گمشده</h3>

<h3 align="right"  style="color: #ff6b35">Income پر کردن مقادیر گمشده در ستون</h3>

```python
clean_data4["Income"] = clean_data4["Income"].fillna(clean_data4.groupby("AgeGroup")["Income"].transform("median"))
print(f"تعداد مقادیر خالی در درآمد: {clean_data4['Income'].isnull().sum()}")
```

<h3 align="right"  style="color: #D980FA">رسم نمودار توزیع درآمد بر اساس سن</h3>

```python
plt.figure(figsize=(12,6))
sns.barplot(x="AgeGroup", y="Income", data=clean_data4, hue="AgeGroup", palette="viridis", legend=False, order=labels)
plt.title("Average Income by AgeGroup", fontsize=16, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```
