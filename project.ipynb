<h3 align="right" style="color: #ff6b35">برای ستون های عددی min و max بررسی</h3>

```python
columns = ['CustomerID', 'Age', 'Income', 'AdSpend', 'ClickThroughRate',
        'ConversionRate', 'WebsiteVisits', 'PagesPerVisit', 'TimeOnSite',
        'SocialShares', 'EmailOpens', 'EmailClicks', 'PreviousPurchases',
        'LoyaltyPoints']

max_len = max(len(col) for col in columns)

for col in columns:
    max_value = df[col].max()
    min_value = df[col].min()
    print(f"{col.ljust(max_len)} => max: {str(max_value).ljust(20)} min: {str(min_value).ljust(20)}")
```

<h3 align="right" style="color: #D980FA">پیدا کردن طولانی ترین نام ستون : max_len</h3>
<h3 align="right" style="color: #D980FA">.می‌باشد left-justify (تراز چپ) مخفف ljust تابع</h3>
<hr>
<h3 align="right" style="color: #ff6b35">بررسی مقادیر یکتا در ستون های غیر عددی</h3>

```python
columns = ['Gender', 'CampaignChannel', 'CampaignType',
           'AdvertisingPlatform', 'AdvertisingTool']

max_len = max(len(col) for col in columns)

for col in columns:
    unique_values = df[col].unique()
    print(f"{col.ljust(max_len)} : {unique_values}")
```

<h3 align="right" style="color: #ff6b35">بررسی و چاپ می‌کنیم DataFrame توزیع مقادیر (فراوانی هر دسته) رو برای سه ستون خاص در</h3>

```python
cols = ['Gender', 'CampaignChannel', 'CampaignType']

for col in cols:
    print(df[col].value_counts(dropna=False))
    print(30 * '-')
```

<h3 align="right"  style="color: #D980FA">با این دستور تعداد تکرار هر مقدار منحصرفرد در ستون رو محاسبه میکنیم : value.counts()</h3>

<h3 align="right"  style="color: #D980FA">بطور پیش فرض دستور بالا مقادیر گمشده رو نادیده میگیره پس برای اینکه شمارش بشن  : dropna</h3>
