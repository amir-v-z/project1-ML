<h3 align="right" style="color: #ff6b35">برای ستون های عددی min و max بررسی</h3>

```python
columns = ['CustomerID', 'Age', 'Income', 'AdSpend', 'ClickThroughRate',
        'ConversionRate', 'WebsiteVisits', 'PagesPerVisit', 'TimeOnSite',
        'SocialShares', 'EmailOpens', 'EmailClicks', 'PreviousPurchases',
        'LoyaltyPoints']

max_len = max(len(col) for col in columns)

for col in columns:
    max_value = df[col].max()
    min_value = df[col].min()
    print(f"{col.ljust(max_len)} => max: {str(max_value).ljust(20)} min: {str(min_value).ljust(20)}")
```

<h3 align="right" style="color: #ff6b35">بررسی مقادیر یکتا در ستون های غیر عددی</h3>

<h3 align="right"  style="color: #4cd137">(Intial Data Visualization) تجسم اولیه داده‌ها</h3>

<h3 align="right"  style="color: #ff6b35">(Histogram , Boxplot نمودار های) نمایش گرافیکی از توزیع و پراکندگی ستون‌های عددی</h3>

```python
numeric_cols = ['Age', 'Income', 'AdSpend', 'ClickThroughRate', 'ConversionRate',
                'WebsiteVisits', 'PagesPerVisit', 'TimeOnSite', 'SocialShares',
                'EmailOpens', 'EmailClicks', 'PreviousPurchases', 'LoyaltyPoints']

fig, axes = plt.subplots(13, 2, figsize=(12, 35))
axes = axes.flatten()

plot_idx = 0
for col in numeric_cols:
    
    # Histogram
    sns.histplot(df[col], kde=True, color="#ff6b35", ax=axes[plot_idx])
    plot_idx += 1
    
    # Boxplot
    sns.boxplot(y=df[col], color="#4cd137", ax=axes[plot_idx])
    axes[plot_idx].set_xlabel(col)
    axes[plot_idx].set_ylabel('Count')
    plot_idx += 1

fig.text(0.24, 0.95, 'Histogram', ha='center', va='center', fontsize=18, fontweight='bold')
fig.text(0.80, 0.95, 'Boxplot', ha='center', va='center', fontsize=18, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.show()
```

<h3 align="right"  style="color: #ff6b35">نمایش گرافیکی نمودار میله‌ای</h3>

```python
object_cols = ['Gender', 'CampaignChannel', 'CampaignType', 'Conversion']

fig, axes = plt.subplots(1, 4, figsize=(15, 4))
axes = axes.flatten()

plot_idx = 0
for col in object_cols:
    counts = df[col].value_counts()
    axes[plot_idx].bar(counts.index.astype(str), counts.values, color="#D980FA")
    axes[plot_idx].set_xlabel(col)
    axes[plot_idx].set_ylabel("Count")
    axes[plot_idx].tick_params(axis='x', rotation=90)
    plot_idx += 1

plt.tight_layout()
plt.show()
```

<h3 align="right"  style="color: #4cd137">(Correlation Analysis) تحلیل همبستگی</h3>

<h3 align="right"  style="color: #ff6b35">(Pearson Correlation Matrix) ماتریس همبستگی پیرسون</h3>

```python
num_cols = df.select_dtypes(include=["int","float"])

corr = num_cols.corr(method="pearson")

fig, ax = plt.subplots(1, 1, figsize=(12, 8))

sns.heatmap(corr, annot=True, cmap="GnBu", fmt=".2f", square=True, ax=ax)
ax.set_title("Correlation Matrix", fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

<h3 align="right"  style="color: #4cd137">clean data</h3>

<h3 align="right"  style="color: #ff6b35">حذف ستون‌های غیرضروری</h3>

```python
clean_data = df.drop(["AdvertisingPlatform", "AdvertisingTool"], axis=1)
clean_data.shape
```

<h3 align="right"  style="color: #ff6b35">AdSpend تمیز کردن ستون</h3>

```python
clean_data['AdSpend'] = clean_data['AdSpend'].replace('[\\$,]', '', regex=True).astype(float)
```

<h3 align="right"  style="color: #ff6b35">EmailOpens & EmailClicks نمایش نمودار پراکندگی برای بررسی رابطه بین دو متغیر</h3>

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].scatter(clean_data['EmailOpens'], clean_data['EmailClicks'], color='#4cd137', alpha=0.7)
axes[0].set_title('Before Cleaning', fontsize=16, fontweight='bold')
axes[0].set_xlabel('EmailOpens')
axes[0].set_ylabel('EmailClicks')
axes[0].grid(True, alpha=0.3)

clean_data2 = clean_data.drop(clean_data[(clean_data["EmailOpens"] == 0) & (clean_data["EmailClicks"] != 0)].index)

axes[1].scatter(clean_data2['EmailOpens'], clean_data2['EmailClicks'], color='#4cd137', alpha=0.7)
axes[1].set_title('After Cleaning', fontsize=16, fontweight='bold')
axes[1].set_xlabel('EmailOpens')
axes[1].set_ylabel('EmailClicks')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"تعداد ردیف قبل از تمیزکاری: {len(clean_data)}")
print(f"تعداد ردیف بعد از تمیزکاری: {len(clean_data2)}")
print(f"تعداد ردیف حذف‌شده: {len(clean_data) - len(clean_data2)}")
```

<h3 align="right"  style="color: #ff6b35">رسم باکس‌پلات برای تشخیص پراکندگی، میانه، چارک ها و نقاط پرت ستون های عددی</h3>

```python
numeric_data = clean_data2[['Age', 'Income', 'AdSpend', 'ClickThroughRate', 'ConversionRate',
                            'WebsiteVisits', 'PagesPerVisit', 'TimeOnSite', 'SocialShares',
                            'EmailOpens', 'EmailClicks', 'PreviousPurchases', 'LoyaltyPoints']]

fig, axes = plt.subplots(5, 3, figsize=(16, 16))
axes = axes.flatten()

plot_idx = 0

for col in numeric_data.columns:
    sns.boxplot(y=numeric_data[col], color='#0652DD', ax=axes[plot_idx])
    axes[plot_idx].set_title(f'{col}', fontweight='bold')
    axes[plot_idx].set_ylabel('')
    plot_idx += 1

for j in range(plot_idx, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```
